```python
supplies = ["pen", "notebook"]
supplies.append("eraser")        # add to the end
print(supplies)
```

```python
supplies.insert(1, "pencil")    # add at position 1
print(supplies)                 # ['pen', 'pencil', 'notebook', 'eraser']
```

```python
supplies.remove("pen")   # remove by value
print(supplies)             # ['pencil', 'notebook', 'eraser']
```

```python
fruits = ["mango", "apple", "grapes", "banana"]
print(len(fruits))            # 4  -> how many items
print("grapes" in fruits)      # True
fruits.sort()                 # sorts the list itself (A-Z)
print(fruits)                
```

```python
temperature = ["31", "28", "35", "40", "26"]
for number in temperature:
    print(number, "°C")
```

```python
temperature = ["31", "28", "35", "40", "26"]
print(sorted(temperature))           # [1, 2, 3, 4]  -> returns a NEW sorted list
print(temperature)                   # [4, 1, 3, 2]  -> original unchanged
```
